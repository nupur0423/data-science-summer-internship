```python
import matplotlib.pyplot as plt
plt.plot([1,2,3,4])
plt.title('simple graph')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.show()
```

```python
values=[5,6,3,7,2]
names=["a","b","c","d","e"]
plt.bar(names,values,color="pink")
plt.show()
```

```python
values=[5,6,3,7,2]
names=["a","b","c","d","e"]
plt.barh(names,values,color="red")
plt.show()
```

```python
ages=[2,5,70,40,30,45,50,45,43,40,24,60,7,13,57,18,90,78,15,97,11,81,29]
range=(0,100)
bins=10
plt.hist(ages,bins,range,color='cyan',histtype='bar',edgecolor='black')
plt.xlabel('age')
plt.ylabel('no.of People')
plt.title('population')
```

```python
x=(2,3,5,5,6,8,9,10,9,7,7)
plt.hist(x,bins=10,color='purple',alpha=0.2,edgecolor='black')
plt.show()
```

```python
x=[0,1,2,3,4,5]
y=[0,1,4,9,16,25]
plt.scatter(x,y,s=50,color="violet")
plt.show()
```

```python
x=[0,1,2,3,4,5]
y=[0,1,4,9,16,25]
plt.scatter(x,y,label="stars",s=50,color="violet",marker="*")
plt.show()
```

```python
activities=['eat','sleep','work','play']
slices=[3,7,8,6]
colors=['r','g','y','b']
plt.pie(slices,labels=activities, colors=colors,startangle=90,shadow=True,explode=(0.1,0.1,0.1,0.1),radius=1.5,autopct='%1.1f%%')
plt.legend(loc="lower right")
plt.show()
```
